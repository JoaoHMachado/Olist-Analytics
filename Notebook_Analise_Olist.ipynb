# Crescimento e Receita


```python
# Junta items com orders pra ter preço + data
df_vendas = pd.merge(
    df_items,
    df_orders,
    on='order_id',
    how='inner'
)

# Confere o tamanho
print(f'df_items: {df_items.shape[0]:,} linhas')
print(f'df_orders: {df_orders.shape[0]:,} linhas')
print(f'df_vendas: {df_vendas.shape[0]:,} linhas')

# Vê as primeiras linhas
df_vendas.head()
```

```python
# Verifica os status dos pedidos

df_vendas['order_status'].value_counts()
```

```python
# Valores antes do filtro
total_itens = df_vendas.shape[0]
print(f'Antes de filtrar por tipo Delivered: {total_itens:,} linhas')

# Cria um novo DataFrame só com entregues (preserva o df_vendas original)
df_vendas_entregues = df_vendas[df_vendas['order_status'] == 'delivered']

# Tamanho depois do filtro
itens_entregues = df_vendas_entregues.shape[0]
print(f'Depois de filtrar por tipo Delivered: {itens_entregues:,} linhas')

# Percentual de entrega
percentual_entrega = (itens_entregues / total_itens) * 100
print(f'Percentual de entrega: {percentual_entrega:.2f}%')
```

```python
# Conta os itens cancelados
itens_cancelados = df_vendas[df_vendas['order_status'] == 'canceled'].shape[0]

# Calcula o percentual
percentual_cancelamento = (itens_cancelados / total_itens) * 100

print(f'Itens cancelados: {itens_cancelados:,}')
print(f'Percentual de cancelamento: {percentual_cancelamento:.2f}%')
```

Insight pro relatório

"Do total de itens vendidos no período, 97,8% foram efetivamente entregues, indicando uma operação logística com alta taxa de conclusão. Apenas 0,48% dos itens foram cancelados."

```python
# Antes da conversão - confere o tipo da coluna
print('Tipo ANTES da conversão:', df_vendas_entregues['order_purchase_timestamp'].dtype)

# Converte a coluna pra tipo datetime
df_vendas_entregues['order_purchase_timestamp'] = pd.to_datetime(df_vendas_entregues['order_purchase_timestamp'])

# Depois da conversão - confere de novo
print('Tipo DEPOIS da conversão:', df_vendas_entregues['order_purchase_timestamp'].dtype)
```

```python
# Cria coluna nova com formato ano-mês
df_vendas_entregues['ano_mes'] = df_vendas_entregues['order_purchase_timestamp'].dt.to_period('M')

# Confere o resultado
df_vendas_entregues[['order_purchase_timestamp', 'ano_mes']].head()
```

```python
# Agrupa por mês e calcula as métricas
evolucao = df_vendas_entregues.groupby('ano_mes').agg(
    qtd_pedidos=('order_id', 'nunique'),
    receita_total=('price', 'sum')
).reset_index()

# Calcula ticket médio (receita ÷ pedidos)
evolucao['ticket_medio'] = evolucao['receita_total'] / evolucao['qtd_pedidos']

# Ordena por mês (do mais antigo pro mais novo)
evolucao = evolucao.sort_values('ano_mes').reset_index(drop=True)

# Mostra o resultado
display(evolucao)
```

```python
# Filtra só a partir de janeiro/2017
evolucao_mensal = evolucao[evolucao['ano_mes'] >= '2017-01'].reset_index(drop=True)

display(evolucao_mensal)
```

**Sobre o período analisado:** Optamos por analisar a partir de janeiro/2017, já que os meses anteriores tiveram volumes muito baixos (sendo que setembro e dezembro de 2016 registraram apenas 1 pedido cada).

```python
import plotly.graph_objects as go

# Converte ano_mes pra texto (se ainda não foi convertido)
evolucao_mensal['ano_mes_str'] = evolucao_mensal['ano_mes'].astype(str)

# Cria a figura vazia
fig_pedidos = go.Figure()

# Adiciona a linha de pedidos (azul-marinho)
fig_pedidos.add_trace(
    go.Scatter(
        x=evolucao_mensal['ano_mes_str'],
        y=evolucao_mensal['qtd_pedidos'],
        name='Pedidos',
        mode='lines+markers',
        line=dict(color='#000A53', width=2),
        marker=dict(size=8, symbol='circle')
    )
)

# Configura o layout no mesmo padrão do gráfico combinado
fig_pedidos.update_layout(
    title=dict(
        text='Evolução Mensal de Pedidos (Jan/2017 - Ago/2018)',
        x=0.5,
        xanchor='center'
    ),
    xaxis=dict(title='Mês', tickangle=-45),
    yaxis=dict(
        title=dict(text='Quantidade de Pedidos', font=dict(color='#000A53')),
        tickfont=dict(color='#000A53')
    ),
    template='plotly_white',
    showlegend=False,
    width=1200,
    height=600
)

fig_pedidos.show()
```

A quantidade de pedidos cresceu de forma consistente em 2017, partindo de 750 em janeiro e chegando a 7.289 em novembro. O pico de novembro coincide com a Black Friday. Em 2018, o volume estabilizou na faixa de 6.000 a 7.000 pedidos por mês.

Em 2018, as variações se estabilizaram entre -10% e +10%, indicando que a operação encontrou um patamar de demanda mais constante.

### Top 10 Estados em Receita

Saber quantos clientes tem cada estado é uma parte da história. Mais importante é saber **quanto cada estado gera de receita**. Vamos cruzar a `df_vendas_entregues` com a `df_customers` para descobrir.

```python
# Junta vendas entregues com clientes pra ter estado de cada venda
df_vendas_geo = pd.merge(
    df_vendas_entregues,
    df_customers[['customer_id', 'customer_state']],
    on='customer_id',
    how='inner'
)

# Confere o tamanho
print(f'df_vendas_entregues: {df_vendas_entregues.shape[0]:,} linhas')
print(f'df_vendas_geo: {df_vendas_geo.shape[0]:,} linhas')

df_vendas_geo.head()
```

```python
# Calcula receita, pedidos e ticket médio por estado
receita_por_estado = df_vendas_geo.groupby('customer_state').agg(
    receita_total=('price', 'sum'),
    qtd_pedidos=('order_id', 'nunique')
).reset_index()

# Calcula o ticket médio
receita_por_estado['ticket_medio'] = receita_por_estado['receita_total'] / receita_por_estado['qtd_pedidos']

# Calcula o % de receita
total_receita = receita_por_estado['receita_total'].sum()
receita_por_estado['pct_receita'] = (receita_por_estado['receita_total'] / total_receita * 100).round(2)

# Ordena pela receita total (do maior pro menor)
receita_por_estado = receita_por_estado.sort_values('receita_total', ascending=False).reset_index(drop=True)

# Pega só o top 10
top10_estados = receita_por_estado.head(10).copy()

display(top10_estados)
```

```python
import plotly.graph_objects as go

# Cores em gradiente: top 3 em destaque, resto em tons mais claros
cores = ['#000A53', '#000A53', '#000A53',
         '#3d5a96', '#3d5a96', '#3d5a96',
         '#7ba7d9', '#7ba7d9', '#7ba7d9', '#7ba7d9']

# Cria texto combinado: "R$ X,XM (XX%)"
top10_estados['texto_barra'] = top10_estados.apply(
    lambda row: f"R$ {row['receita_total']/1_000_000:.2f}M ({row['pct_receita']}%)",
    axis=1
)

# Cria o gráfico
fig_estados = go.Figure()

fig_estados.add_trace(
    go.Bar(
        x=top10_estados['receita_total'],
        y=top10_estados['customer_state'],
        orientation='h',
        marker=dict(
            color=cores,
            line=dict(color='#000A53', width=1)
        ),
        text=top10_estados['texto_barra'],
        textposition='outside',
        textfont=dict(size=12, color='#000A53')
    )
)

# Configura o layout no padrão do relatório
fig_estados.update_layout(
    title=dict(
        text='<b>Top 10 Estados em Receita</b>',
        x=0.5,
        xanchor='center',
        font=dict(size=18)
    ),
    xaxis=dict(
        title='Receita Total (R$)',
        tickprefix='R$ ',
        tickformat=',.0f',
        showgrid=True,
        gridcolor='#e5e5e5',
        range=[0, 6500000]
    ),
    yaxis=dict(
        title='Estado',
        autorange='reversed',
        tickfont=dict(size=13, color='#000A53')
    ),
    template='plotly_white',
    showlegend=False,
    width=1200,
    height=600,
    margin=dict(l=80, r=200, t=100, b=60)
)

fig_estados.show()
```

**Top 10 estados em receita:**

São Paulo lidera com folga: representa **38% da receita total** da Olist, mais que a soma dos três próximos estados juntos (RJ + MG + RS = 30,56%). O Sudeste como um todo (SP + RJ + MG + ES) concentra cerca de **65% da receita**, evidenciando alta dependência geográfica.

Vale destacar que **a Bahia tem o maior ticket médio** entre os top 10 (R$ 151,59), maior até que o de São Paulo (R$ 125,12). Isso sugere que clientes do Nordeste, embora em menor volume, fazem compras de maior valor por pedido — possivelmente porque o frete e o acesso a produtos são mais limitados na região, levando a compras mais "concentradas".

Estados do Nordeste (BA, PE, CE) aparecem em posições discretas no ranking, e nenhum estado do Norte aparece no top 10, sinalizando regiões com **menor penetração** e potencial oportunidade de expansão.

### Distribuição por Região

Distribuição de receita entre Norte, Nordeste, Centro-Oeste, Sudeste e Sul.

```python
# Mapeia cada estado para sua região
regioes = {
    'AC': 'Norte', 'AP': 'Norte', 'AM': 'Norte', 'PA': 'Norte',
    'RO': 'Norte', 'RR': 'Norte', 'TO': 'Norte',
    'AL': 'Nordeste', 'BA': 'Nordeste', 'CE': 'Nordeste', 'MA': 'Nordeste',
    'PB': 'Nordeste', 'PE': 'Nordeste', 'PI': 'Nordeste', 'RN': 'Nordeste', 'SE': 'Nordeste',
    'DF': 'Centro-Oeste', 'GO': 'Centro-Oeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste',
    'ES': 'Sudeste', 'MG': 'Sudeste', 'RJ': 'Sudeste', 'SP': 'Sudeste',
    'PR': 'Sul', 'RS': 'Sul', 'SC': 'Sul'
}

# Cria a coluna de região a partir do estado
df_vendas_geo['regiao'] = df_vendas_geo['customer_state'].map(regioes)

# Confere se criou certo
df_vendas_geo[['customer_state', 'regiao']].head()
```

```python
# Calcula receita, pedidos e clientes únicos por região
receita_por_regiao = df_vendas_geo.groupby('regiao').agg(
    receita_total=('price', 'sum'),
    qtd_pedidos=('order_id', 'nunique'),
    qtd_clientes=('customer_id', 'nunique')
).reset_index()

# Calcula o ticket médio
receita_por_regiao['ticket_medio'] = receita_por_regiao['receita_total'] / receita_por_regiao['qtd_pedidos']

# Calcula percentual de receita
total_receita = receita_por_regiao['receita_total'].sum()
receita_por_regiao['pct_receita'] = (receita_por_regiao['receita_total'] / total_receita * 100).round(2)

# Arredonda o ticket médio
receita_por_regiao['ticket_medio'] = receita_por_regiao['ticket_medio'].round(2)

# Ordena pela receita (do maior pro menor)
receita_por_regiao = receita_por_regiao.sort_values('receita_total', ascending=False).reset_index(drop=True)

display(receita_por_regiao)
```

```python
import plotly.graph_objects as go

# Gráfico de barras: distribuição de receita por região
fig_regiao = go.Figure()

fig_regiao.add_trace(
    go.Bar(
        x=receita_por_regiao['regiao'],
        y=receita_por_regiao['receita_total'],
        marker=dict(color='#000A53'),
        text=receita_por_regiao['pct_receita'],
        texttemplate='%{text}%',
        textposition='outside'
    )
)

# Configura o layout no padrão do relatório
fig_regiao.update_layout(
    title=dict(
        text='<b>Distribuição de Receita por Região</b>',
        x=0.5,
        xanchor='center',
        font=dict(size=18)
    ),
    xaxis=dict(title='Região'),
    yaxis=dict(
        title=dict(text='Receita Total (R$)', font=dict(color='#000A53')),
        tickfont=dict(color='#000A53'),
        tickprefix='R$ ',
        tickformat=',.0f'
    ),
    template='plotly_white',
    showlegend=False,
    width=1200,
    height=600
)

fig_regiao.show()
```

**Distribuição por região:**

A operação da Olist é altamente concentrada no Sudeste, que responde por **65% da receita** total. Sul aparece em segundo lugar com 14%, seguido por Nordeste (11%), Centro-Oeste (6%) e Norte (apenas 2%).

Em outras palavras: **2 em cada 3 reais faturados pela Olist vêm de SP, RJ, MG ou ES**. As regiões Norte e Nordeste juntas representam pouco mais de 13% do negócio, indicando uma das maiores oportunidades estratégicas de expansão para diversificar o risco geográfico.

### Ticket Médio por Região



```python
# Gráfico de barras: ticket médio por região
fig_ticket_regiao = go.Figure()

fig_ticket_regiao.add_trace(
    go.Bar(
        x=receita_por_regiao['regiao'],
        y=receita_por_regiao['ticket_medio'],
        marker=dict(color='#000A53'),
        text=receita_por_regiao['ticket_medio'].round(0),
        texttemplate='R$ %{text}',
        textposition='outside'
    )
)

# Configura o layout no padrão do relatório
fig_ticket_regiao.update_layout(
    title=dict(
        text='<b>Ticket Médio por Região</b>',
        x=0.5,
        xanchor='center',
        font=dict(size=18)
    ),
    xaxis=dict(title='Região'),
    yaxis=dict(
        title=dict(text='Ticket Médio (R$)', font=dict(color='#000A53')),
        tickfont=dict(color='#000A53'),
        tickprefix='R$ ',
        tickformat=',.0f'
    ),
    template='plotly_white',
    showlegend=False,
    width=1200,
    height=600
)

fig_ticket_regiao.show()
```

**Ticket médio por região:**

Um padrão interessante aparece quando olhamos o ticket médio: regiões com **menor volume** apresentam **ticket médio maior**. O Norte, que representa apenas 2,47% da receita total, tem o maior ticket médio (R$ 182,11), enquanto o Sudeste, com 65% da receita, tem o menor ticket (R$ 130,64) — uma diferença de quase 40%.

Esse comportamento sugere dois perfis de consumo distintos:

- **Sudeste:** muitos clientes comprando com frequência, mas em valores menores (mercado maduro, acesso fácil a produtos)
- **Norte/Nordeste:** poucos clientes, porém com compras de maior valor por pedido (acesso limitado leva a compras mais "concentradas")

Para estratégias de CRM, essa distinção é fundamental: o Sudeste pede campanhas de **frequência** e o Norte/Nordeste, campanhas de **expansão de base** com tickets já naturalmente maiores.

### Conclusão da Análise Geográfica

A análise revelou três descobertas geográficas importantes:

1. **Alta concentração no Sudeste:** quatro estados (SP, RJ, MG, ES) respondem por 65% da receita da Olist, sendo que São Paulo sozinho representa 38%. Isso configura um **risco de concentração geográfica** relevante para o negócio.

3. **Ticket médio comporta-se inversamente ao volume:** regiões com menor volume apresentam ticket médio maior (Norte: R$ 182 vs Sudeste: R$ 131). Isso indica que clientes do Norte e Nordeste já chegam com **maior propensão a gastar por pedido** — só falta volume e capilaridade.

**Implicação estratégica:** diversificar geograficamente a base de clientes da Olist é uma algo importante para o crescimento sustentável e de redução de risco. Investir em logística e parcerias regionais para Nordeste e Norte pode acelerar o crescimento explorando clientes que já demonstram alto valor por compra.

```python
# Função pra agrupar parcelas em faixas
def agrupar_parcelas(num):
    if num == 1:
        return '1. À vista'
    elif num <= 4:
        return '2. Curto (2-4x)'
    elif num <= 10:
        return '3. Médio (5-10x)'
    else:
        return '4. Longo (acima de 10x)'

# Aplica a função na coluna de parcelas
df_credito['faixa_parcela'] = df_credito['payment_installments'].apply(agrupar_parcelas)

# Conta quantos pagamentos em cada faixa
faixas = df_credito['faixa_parcela'].value_counts().reset_index()
faixas.columns = ['faixa', 'quantidade']

# Calcula o percentual
total_credito = faixas['quantidade'].sum()
faixas['percentual'] = (faixas['quantidade'] / total_credito * 100).round(2)

# Ordena alfabeticamente (por isso usei números no nome da faixa)
faixas = faixas.sort_values('faixa').reset_index(drop=True)

display(faixas)
```

```python
# Calcula o ticket médio por faixa de parcelas
ticket_por_faixa = df_credito.groupby('faixa_parcela').agg(
    ticket_medio=('payment_value', 'mean'),
    quantidade=('payment_value', 'count')
).reset_index()

# Renomeia a coluna pra ficar mais clara
ticket_por_faixa = ticket_por_faixa.rename(columns={'faixa_parcela': 'faixa'})

# Arredonda o ticket médio
ticket_por_faixa['ticket_medio'] = ticket_por_faixa['ticket_medio'].round(2)

# Ordena pelas faixas (1, 2, 3, 4)
ticket_por_faixa = ticket_por_faixa.sort_values('faixa').reset_index(drop=True)

display(ticket_por_faixa)
```

```python
# Filtra só pedidos entregues (que têm data de entrega real)
df_entregues = df_pedidos_reviews[df_pedidos_reviews['order_status'] == 'delivered'].copy()

# Converte as datas pra tipo datetime
df_entregues['order_delivered_customer_date'] = pd.to_datetime(df_entregues['order_delivered_customer_date'])
df_entregues['order_estimated_delivery_date'] = pd.to_datetime(df_entregues['order_estimated_delivery_date'])

# Calcula a diferença em dias entre entrega real e estimada
# Negativo = adiantado / Positivo = atrasado
df_entregues['dias_atraso'] = (df_entregues['order_delivered_customer_date'] - df_entregues['order_estimated_delivery_date']).dt.days

# Confere o tamanho
print(f'Pedidos entregues com review: {df_entregues.shape[0]:,}')

# Mostra as primeiras linhas pra conferir
df_entregues[['order_id', 'order_estimated_delivery_date', 'order_delivered_customer_date', 'dias_atraso', 'review_score']].head()
```

```python
# Função pra agrupar em faixas de atraso
def faixa_atraso(dias):
    if dias <= -10:
        return '1. Muito adiantado (10+ dias antes)'
    elif dias <= -1:
        return '2. Adiantado (1-10 dias antes)'
    elif dias == 0:
        return '3. No prazo'
    elif dias <= 10:
        return '4. Atrasado (1-10 dias depois)'
    else:
        return '5. Muito atrasado (10+ dias depois)'

# Aplica a função na coluna de dias_atraso
df_entregues['faixa_entrega'] = df_entregues['dias_atraso'].apply(faixa_atraso)

# Calcula a nota média e quantidade por faixa
nota_por_faixa = df_entregues.groupby('faixa_entrega').agg(
    nota_media=('review_score', 'mean'),
    quantidade=('review_score', 'count')
).reset_index()

# Calcula percentual
total = nota_por_faixa['quantidade'].sum()
nota_por_faixa['percentual'] = (nota_por_faixa['quantidade'] / total * 100).round(2)

# Ordena pelas faixas (1, 2, 3, 4, 5)
nota_por_faixa = nota_por_faixa.sort_values('faixa_entrega').reset_index(drop=True)

# Arredonda a nota média pra 2 casas
nota_por_faixa['nota_media'] = nota_por_faixa['nota_media'].round(2)

display(nota_por_faixa)
```

```python
# Vamos ver a % de pedidos atrasados por mês
df_entregues['ano_mes_entrega'] = pd.to_datetime(df_entregues['order_purchase_timestamp']).dt.to_period('M')

# Cria coluna que marca se foi atraso (dias > 0)
df_entregues['atrasou'] = df_entregues['dias_atraso'] > 0

# Calcula % de atraso por mês
atraso_por_mes = df_entregues.groupby('ano_mes_entrega').agg(
    total_pedidos=('order_id', 'count'),
    pedidos_atrasados=('atrasou', 'sum')
).reset_index()

atraso_por_mes['pct_atraso'] = (atraso_por_mes['pedidos_atrasados'] / atraso_por_mes['total_pedidos'] * 100).round(2)

# Filtra a partir de jan/2017 e mostra
atraso_por_mes = atraso_por_mes[atraso_por_mes['ano_mes_entrega'] >= '2017-01'].reset_index(drop=True)

display(atraso_por_mes)
```

##  Retenção e Recompra

Crescer em receita e ter alta satisfação é importante, mas há uma métrica ainda mais decisiva para o e-commerce: **o cliente volta a comprar?** Operações que dependem de aquisição constante de novos clientes são frágeis — toda parada de investimento em marketing derruba a receita. Esta análise investiga a base de clientes da Olist para entender o comportamento de recompra e o potencial de lifetime value.

### O dataset distingue cliente de pedido

Antes de calcular qualquer coisa, é importante entender uma sutileza importante da tabela `df_customers`:

- **`customer_id`** → identificador único **por pedido** (cada compra gera um novo)
- **`customer_unique_id`** → identificador único **da pessoa** (mesmo em todas as compras dela)

```python
# Conta pedidos vs clientes únicos
total_pedidos = df_customers['customer_id'].nunique()
total_clientes_unicos = df_customers['customer_unique_id'].nunique()

print(f'Total de pedidos:                      {total_pedidos:,}')
print(f'Total de clientes únicos (pessoas):    {total_clientes_unicos:,}')
print(f'Pedidos a mais que clientes:           {total_pedidos - total_clientes_unicos:,}')
print()
print(f'Taxa de recompra (em pedidos):         {((total_pedidos - total_clientes_unicos) / total_pedidos * 100):.2f}%')
```

```python
# Conta quantos pedidos cada cliente único fez
pedidos_por_cliente = df_customers.groupby('customer_unique_id')['customer_id'].nunique().reset_index()
pedidos_por_cliente.columns = ['customer_unique_id', 'qtd_pedidos']

# Tabela completa de distribuição (documentação)
distribuicao_completa = pedidos_por_cliente['qtd_pedidos'].value_counts().reset_index()
distribuicao_completa.columns = ['num_compras', 'qtd_clientes']
distribuicao_completa = distribuicao_completa.sort_values('num_compras').reset_index(drop=True)

# Percentual
total = distribuicao_completa['qtd_clientes'].sum()
distribuicao_completa['percentual'] = (distribuicao_completa['qtd_clientes'] / total * 100).round(2)

# Percentual acumulado
distribuicao_completa['pct_acumulado'] = distribuicao_completa['percentual'].cumsum().round(2)

display(distribuicao_completa)
```

```python
# Agrupa em categorias executivas (abrindo até 6, agrupando 7+)
def faixa_recompra(num):
    if num == 1:
        return '1 compra'
    elif num == 2:
        return '2 compras'
    elif num == 3:
        return '3 compras'
    elif num == 4:
        return '4 compras'
    elif num == 5:
        return '5 compras'
    elif num == 6:
        return '6 compras'
    else:
        return '7+ compras'

pedidos_por_cliente['faixa_recompra'] = pedidos_por_cliente['qtd_pedidos'].apply(faixa_recompra)

# Conta clientes por faixa
recompra_agrupada = pedidos_por_cliente['faixa_recompra'].value_counts().reset_index()
recompra_agrupada.columns = ['faixa', 'qtd_clientes']

# Calcula percentual
total = recompra_agrupada['qtd_clientes'].sum()
recompra_agrupada['percentual'] = (recompra_agrupada['qtd_clientes'] / total * 100).round(2)

# Ordena na ordem lógica
ordem = ['1 compra', '2 compras', '3 compras', '4 compras', '5 compras', '6 compras', '7+ compras']
recompra_agrupada['faixa'] = pd.Categorical(recompra_agrupada['faixa'], categories=ordem, ordered=True)
recompra_agrupada = recompra_agrupada.sort_values('faixa').reset_index(drop=True)

display(recompra_agrupada)
```

```python
import plotly.graph_objects as go

# Gráfico de barras: distribuição de recompra
fig_recompra = go.Figure()

fig_recompra.add_trace(
    go.Bar(
        x=recompra_agrupada['faixa'].astype(str),
        y=recompra_agrupada['qtd_clientes'],
        marker=dict(color='#000A53'),
        text=recompra_agrupada['percentual'],
        texttemplate='%{text}%',
        textposition='outside'
    )
)

fig_recompra.update_layout(
    title=dict(
        text='<b>Distribuição de Clientes por Número de Compras</b>',
        x=0.5,
        xanchor='center',
        font=dict(size=18)
    ),
    xaxis=dict(title='Número de Compras'),
    yaxis=dict(
        title=dict(text='Quantidade de Clientes', font=dict(color='#000A53')),
        tickfont=dict(color='#000A53'),
        tickformat=',.0f'
    ),
    template='plotly_white',
    showlegend=False,
    width=1200,
    height=600
)

fig_recompra.show()
```

**Retenção: o principal ponto de atenção da operação**

A análise da base de clientes revela o achado mais crítico de todo o relatório: a Olist é eficiente em atrair clientes, mas falha em retê-los. Dos quase 96 mil clientes únicos, a grande maioria realizou apenas uma compra. Somando quem comprou uma ou duas vezes, chega-se a praticamente toda a base.

A recompra é rara: clientes com três ou mais compras representam uma fração mínima do total. O cliente mais fiel da base realizou 17 compras — um caso isolado, provavelmente um pequeno revendedor que utiliza a plataforma de forma recorrente.

Esse padrão tem implicação direta para o negócio: todo o crescimento observado nas trilhas anteriores foi sustentado por aquisição de novos clientes, não por recompra da base existente. Em um cenário de aumento do custo de aquisição (CAC) ou desaceleração do investimento em marketing, a receita ficaria vulnerável.

Vale uma ressalva metodológica: por se tratar de um marketplace que opera através de outras plataformas (como Mercado Livre), parte da recompra pode não ser capturada — o mesmo cliente pode aparecer com identificadores diferentes. Ainda assim, o padrão geral de baixa retenção é consistente e merece atenção estratégica.

##  Trilha 5: Oportunidades e Recomendações

### Síntese das descobertas

A análise do dataset da Olist revelou uma operação em forte crescimento, com desafios pontuais que merecem atenção estratégica:

**Crescimento e Receita**
- Crescimento de 10x em pedidos e 9x em receita entre janeiro e novembro de 2017
- Estabilização em 2018 no patamar de 6.000 a 7.000 pedidos por mês
- Ticket médio mantido estável (R$ 124 a R$ 149), indicando que o crescimento veio do volume de clientes
- Forte concentração geográfica: o Sudeste responde por 65% da receita, enquanto Norte e Nordeste somam apenas 13,8% — apesar de apresentarem os maiores tickets médios (até R$ 182)

**Mix de Produtos**
- As 10 maiores categorias concentram 63% da receita, lideradas por Beleza & Saúde e Relógios & Presentes
- Categorias com lógicas distintas: algumas faturam por volume (ticket baixo), outras por valor agregado (ticket alto)

**Comportamento e Pagamentos**
- Cartão de crédito domina as transações (74%), seguido pelo boleto (19%)
- O brasileiro prefere parcelamentos curtos: 72% pagam à vista ou em até 4x
- Quanto maior o parcelamento, maior o ticket — clientes que parcelam em 10x gastam 4x mais que os que pagam à vista

**Satisfação e Retenção**
- 77% dos clientes dão notas 4 ou 5, mas 11,5% dão nota mínima
- 92% das entregas chegam antes do prazo; quando há atraso, a nota despenca de 4,03 para 2,53 (queda de 37%)
- Apenas 3% dos clientes voltam a comprar — a retenção é o principal ponto de atenção da operação

### Principais riscos identificados

A partir das análises, identificamos cinco riscos que merecem atenção estratégica da Olist:

**1. Baixa retenção de clientes**

Apenas 3% dos clientes realizam uma segunda compra. Todo o crescimento observado foi sustentado por aquisição de novos clientes, não por recompra da base existente. Esse é o risco mais relevante do negócio: em um cenário de aumento do custo de aquisição (CAC) ou desaceleração do investimento em marketing, a receita ficaria diretamente exposta.

**2. Vulnerabilidade logística em picos de demanda**

Os meses de maior volume (novembro/2017, fevereiro/2018 e março/2018) foram também os de maior taxa de atraso, chegando a 18,6% em março/2018. Como a satisfação é diretamente impactada por atrasos (queda de 37% na nota média), o crescimento futuro pode ser comprometido se a capacidade logística não escalar junto com o volume.

**3. Concentração geográfica**

O Sudeste concentra 65% da receita, com São Paulo respondendo sozinho por 38%. Essa dependência de uma única região representa risco em caso de saturação do mercado ou eventos regionais adversos. Ao mesmo tempo, é uma oportunidade: Norte e Nordeste somam apenas 13,8% da receita, mas têm os maiores tickets médios do país.

**4. Polarização das avaliações**

Apesar de 77% dos clientes avaliarem positivamente, 11,5% dão nota mínima — mais de 11 mil clientes muito insatisfeitos em uma base de quase 100 mil pedidos. Em mercados onde reputação online influencia decisões de compra, essa cauda longa de insatisfação pode prejudicar a aquisição de novos clientes.

**5. Dependência de um único canal de pagamento**

74% das transações dependem do cartão de crédito. Mudanças regulatórias, taxas de intercâmbio ou problemas com bandeiras impactariam três quartos da operação. A diversificação de meios de pagamento (como o Pix, indisponível no período do dataset) reduz esse risco.

### Recomendações estratégicas

A partir dos riscos identificados, propomos cinco recomendações prioritárias:

**1. Construir uma estratégia de retenção e recompra**

Com apenas 3% de recompra, há um espaço enorme a ser explorado. Ações possíveis: programa de fidelidade com benefícios crescentes, e-mail marketing pós-compra com recomendações personalizadas, cupons para a segunda compra e réguas de relacionamento segmentadas. Aumentar a recompra é a alavanca de maior impacto sobre a receita, já que reter custa menos que adquirir.

**2. Investir em capacidade logística escalável**

Como atrasos são o principal driver de insatisfação, a operação logística precisa escalar junto com o crescimento. Isso inclui ampliar parcerias com transportadoras, estabelecer hubs regionais e criar planos de contingência para datas críticas como a Black Friday. Manter a taxa de atraso abaixo de 5% deve ser uma meta operacional.

**3. Explorar a expansão para Norte e Nordeste**

Essas regiões representam apenas 13,8% da receita, mas têm os maiores tickets médios do país (até R$ 182). Investir em logística regional, parcerias locais e capilaridade pode destravar crescimento com receita unitária superior, ao mesmo tempo em que reduz a dependência do Sudeste.

**4. Aumentar o ticket médio com estratégias por categoria**

Como o crescimento veio de volume e não de ticket, vale diversificar a receita por cliente explorando o perfil de cada categoria. Categorias de ticket alto (Relógios & Presentes, Cool Stuff) são candidatas a ações de destaque e cross-sell; categorias de alto volume e ticket baixo (Cama, Mesa & Banho) pedem kits e combos para elevar o valor por pedido. Frete grátis acima de um valor mínimo também estimula carrinhos maiores.

**5. Diversificar meios de pagamento**

Reduzir a dependência do cartão de crédito (74%) ampliando alternativas como Pix, carteiras digitais e parcelamento via boleto. Além de reduzir riscos operacionais, isso diminui custos com taxas de intercâmbio.

### Conclusão final

A Olist apresentou um período de crescimento expressivo entre 2017 e 2018, com indicadores comerciais saudáveis e boa satisfação geral. A operação demonstrou capacidade de execução em escala, mas a análise revelou que esse crescimento foi sustentado quase inteiramente por aquisição de novos clientes — a recompra, de apenas 3%, é o principal ponto de atenção.

Os riscos identificados se concentram em cinco frentes endereçáveis: baixa retenção, capacidade logística sob pressão, concentração geográfica, polarização das avaliações e dependência do cartão de crédito. Nenhum deles é estrutural.

A boa notícia para investidores é que cada risco tem como contrapartida uma oportunidade clara: a baixa retenção é espaço para fidelização; a concentração no Sudeste convive com regiões de alto ticket ainda inexploradas; e o mix de categorias abre caminho para estratégias de cross-sell. Trata-se de uma operação fundamentalmente bem estruturada, com oportunidade de crescimento identificáveis e acionáveis.
